```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

pos_list = ['rb', 'qb', 'wr', 'te']

df_list = []

errors_list = []

full_column_list = [
    'PLAYER', 'POS',
    'RECEIVING_REC', 'RECEIVING_YDS', 'RECEIVING_TDS', 'RUSHING_ATT', 'RUSHING_YDS', 'RUSHING_TDS', #WR/RB
    'PASSING_ATT', 'PASSING_CMP', 'PASSING_YDS', 'PASSING_TDS', 'PASSING_INTS', #passing
    'FUMBLES', 'PPR_SCORING'
    ]

for pos in pos_list:

    print(f'Grabbing data for {pos.upper()} ...')

    url = f'https://www.fantasypros.com/nfl/projections/{pos}.php?week=draft&scoring=PPR&week=draft'

    r = requests.get(url)

    soup = BeautifulSoup(r.content, 'html.parser')

    table = soup.find_all('table', attrs={'id':'data'})

    tdf = pd.read_html(str(table))[0]

    #joins the multi-index to create unique column name
    tdf.columns = ['_'.join(col) for col in tdf.columns]
    tdf.rename(columns={
                'Unnamed: 0_level_0_Player': 'PLAYER',
                'MISC_FL' : 'FUMBLES',
                'MISC_FPTS' : 'PPR_SCORING'
                }, inplace=True)

    tdf['POS'] = pos.upper()
    
    tdf = tdf.reindex(columns = full_column_list, fill_value = 0)

    tdf['std_scoring'] = tdf['PPR_SCORING'] - tdf['RECEIVING_REC']
    tdf['half_ppr_scoring'] = tdf['PPR_SCORING'] - 0.5*tdf['RECEIVING_REC']

    tdf.columns = [col.lower() for col in tdf.columns]

    df_list.append(tdf)

df = pd.concat(df_list)

df.head()
```
